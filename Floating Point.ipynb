**PROBLEMA 7**


```
#Se establecen los valores iniciales de la recursion
x1 = 1/3
x2 = 1/12
print('x1=',frac(x1).limit_denominator(),'\nx2=', frac(1/12).limit_denominator(), sep='')

# Se genera la lista que imprime el valor de la recurrencia y al frente su valor en forma de fraccion
z = [1/3, 1/12]
exac = [1/3, 1/12]
for i in range (3,80):
  xi = 2.25*x2 - 0.5*x1
  print('x',i,'=',"{0:.16f}".format(xi),'\t \t', frac(xi).limit_denominator())
  x1 = x2
  x2 = xi
  z.append(xi)
  exac.append((1/3)*(1/4)**i)
  
# Se genera un grafica que permita ver como cambia la recursion en el momento en que diverge
x=list(range(1,80))
plt.figure(figsize=(10, 5))
plt.plot(x, exac, 'r-', label = 'Anal√≠tica')
plt.plot(x, z, 'b-', label = 'Recurrencia')
plt.axis([30, 80, -1, 5])
legend = plt.legend(loc='upper left', shadow=True, fontsize='x-large')
plt.show()
```
